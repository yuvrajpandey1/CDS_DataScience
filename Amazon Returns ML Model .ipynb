```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('sales_returns.csv')
sales
```

```python
# To display first five records
sales.head()
```

```python
# To see structure of data , total number of rows and column present 
sales.info()
```

```python
# Filtering and taking only 4 fields for further analysis
X = sales.loc[:,['cust_age','cust_gender','city','sub_category']]
y = sales.returned
X.head()
```

```python
# Converting text data to number because machine learning understand number better
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
X.cust_gender = enc.fit_transform(X.cust_gender)
X.city = enc.fit_transform(X.city)
X.sub_category = enc.fit_transform(X.sub_category)
X.head()
```

```python
# Using mL model 
from sklearn.ensemble import RandomForestClassifier
mod = RandomForestClassifier()
mod.fit(X,y)
```

```python
mod
```

```python
y
```

```python
# Checking by taking random customer data
mod.predict([[27,1,100,10]])
```

* So False means this order will not returned.

```python
mod.predict([[25,0,90,15]])
```

```python
mod.predict([[22,1,100,8]])
```

* Might be this order will return

```python
# Checking accuracy of model
mod.score(X,y)
```

* Getting 99% of accuracy
