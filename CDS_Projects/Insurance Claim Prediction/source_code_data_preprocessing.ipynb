Spliting the data set into categorical and numerical and boolean form 


```python
#Create a frame that stores multiple data 
data = []
for f in ins_data.columns:
    #Define the role
    if f == 'target':
        role = 'target'
    elif f =='id':
        role = 'id'
    else:
        role = 'input'
    
    #Define the type of data 
    if 'bin' in f or f =='target':
        type_of_data = 'binary'
    elif 'cat' in f or f == 'id':
        type_of_data = 'nominal'
    elif ins_data[f].dtype == 'float64':
        type_of_data = 'continuous'
    elif ins_data[f].dtype == 'int64':
        type_of_data = 'ordinal'
        
        #Define the keep
    keep = True
    if f =='id':
        keep = False
    
    #Defining the dtype
    dtype = ins_data[f].dtype
    type_of_data
    #Creating a Dict that contains all the metadata for the variable 
    f_dict = {
        'varname': f,
        'role': role,
        'type_of_data': type_of_data,
        'keep': keep,
        'dtype': dtype
    }
    data.append(f_dict)
    
meta = pd.DataFrame(data, columns = ['varname', 'role', 'type_of_data', 'keep', 'dtype'])
meta.set_index('varname', inplace=True)
```
