### Model 

* spliting data for solo matches 

```python
df = pubg[pubg['matchType']=='solo']
df.head()
df1 = pubg[pubg['matchType']=='solo-fpp']
print(df1.shape)
df2 = pubg[pubg['matchType']=='normal-solo-fpp']
print(df2.shape)
df3 = pubg[pubg['matchType']=='normal-solo']
print(df3.shape)
solo = pd.concat([df,df1,df2,df3],axis=0)
print(solo.shape)
```

```python
for m in ['Id', 'groupId', 'matchId']:
    print(f'unique[{m}] count: ',solo[m].nunique())
```

```python
solo.corrwith(solo.winPlacePerc)
```

* DBNO and Revives are irrelevent for solo matches so we ll drop these features for solo matches  

```python
solo.drop(['DBNOs','revives'],axis=1,inplace=True)
```

```python
solo.corrwith(solo.winPlacePerc)
```

* spliting data for dua matches 

```python
d1 = pubg[pubg['matchType']=='duo']
print('duo : ',d1.shape)
d2 = pubg[pubg['matchType']=='duo-fpp']
print('duo-fpp : ',d2.shape)
d3 = pubg[pubg['matchType']=='normal-duo']
print('normal-duo : ',d3.shape)
d4 = pubg[pubg['matchType']=='normal-duo-fpp']
print('normal-duo-fpp : ',d4.shape)
d5 = pubg[pubg['matchType']=='crashfpp']
print('crashfpp : ',d5.shape)
d6 = pubg[pubg['matchType']=='crashtpp']
print('crashtpp : ',d6.shape)
duo = pd.concat([d1,d2,d3,d4,d5,d6],axis=0)
print('total duo : ',duo.shape)
```

```python
for m in ['Id', 'groupId', 'matchId']:
    print(f'unique[{m}] count: ',duo[m].nunique())
```

```python
duo.corrwith(duo.winPlacePerc)
```

* spliting data for squad matches 

```python
s1 = pubg[pubg['matchType']=='squad']
print('squad : ',s1.shape)
s2 = pubg[pubg['matchType']=='squad-fpp']
print('squad-fpp : ',s2.shape)
s3 = pubg[pubg['matchType']=='normal-squad']
print('normal-squad : ',s3.shape)
s4 = pubg[pubg['matchType']=='normal-squad-fpp']
print('normal-squad-fpp : ',s4.shape)
s5 = pubg[pubg['matchType']=='flarefpp']
print('flarefpp : ',s5.shape)
s6 = pubg[pubg['matchType']=='flaretpp']
print('flaretpp : ',s6.shape)
sqd = pd.concat([s1,s2,s3,s4,s5,s6],axis=0)
print('total sqd : ',sqd.shape)
```

```python
for m in ['Id', 'groupId', 'matchId']:
    print(f'unique[{m}] count: ',sqd[m].nunique())
```

```python
sqd.corrwith(sqd.winPlacePerc)
```
