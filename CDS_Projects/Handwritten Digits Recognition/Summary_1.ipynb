## HandwrittenDigits - Classification of given image of a handwritten digit into one of 10 classes representing integer values from 0 to 9, inclusively.

## 1. Project summary

The MNIST dataset is an acronym that stands for the Modified National Institute of Standards and Technology dataset.
It is a dataset of 60,000 small square 28×28 pixel grayscale images of handwritten single digits between 0 and 9. Our target here to create a ML model to classify the image of a handwritten digit into one of 10 classes representing integer values from 0 to 9, inclusively.

The input file is a image dataset. The shape of the data is 60,000x28x28. The dataset already has a well-defined train and test dataset that we can use.
 
The MNIST dataset contains 60,000 training cases and 10,000 test cases of handwritten digits (0 to 9). Each digit is normalized and centered in a gray-scale (0 - 255) image with size 28 × 28. Each image consists of 784 pixels that represent the features of the digits.

Goal: Classification of hand written digits into their respective category.

The models used are SVM, Random Forest,KNN, Tensorflow Keras dense layer and along with CNN.

CNN with tensorflow gives the best result of Error : 0.0507 and accuracy : 0.984 for the whole data set.

Various steps such as Preprocessing, Encoding , prediction using Machine Learning models are used to achieve our objective.
Visualization is done by Matplotlib and Seaborn.
