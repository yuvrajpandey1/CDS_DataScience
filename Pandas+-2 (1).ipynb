```python
## Handle and fill the nan value
#Methods--> fill the data-->any value
### ffill,bfill,mean,median,mode(Categorical Data)
```

```python
cars.fillna(method='bfill')   #fills with previous value(LOCF)
```

### Grouping --> groupby and use any aggregate function

```python
count() - Number of non-null observations.
sum()  - Sum of values
mean() - Mean of calues
median()- Arithmetic median of values
min()  - Minimum
max()  - Maximum
mode() - Mode
std()  - Standard deviation
var()  - Variance
```
