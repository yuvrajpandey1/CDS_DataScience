### Object Oriented Plotting
4 Steps
* Create a blank figure object
* Add Axes
* Generate plots with the object
* specify layout & plotting parameters

```python
fig = plt.figure()
ax = fig.add_axes([0.8,0.5,0.9,1])  # L,B,W,H
ax.plot(x,y)
plt.show()
```

```python
## Setting Limits and ticks
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  #L,B,W,H
# ax.set_xlim([1,3])
ax.set_ylim([15,65])
# ax.set_xticks([1,3,5])
ax.set_yticks([9,25])
ax.plot(x,y)
plt.show()
```

 ### Multiple axes Matplotlib with different scales

```python
import matplotlib.pyplot as plt
fig,ax1 = plt.subplots()
ax1.plot([1,2,3,4,5],[3,5,7,1,9], color='red')
ax2 = ax1.twinx()
ax2.plot([11,12,31,41,15],[13,51,17,11,76], color='blue')
fig.tight_layout()
plt.show()
```

### Generating multiple plots


```python
fig = plt.figure(figsize=(15,15))
fig, ax = plt.subplots(2,3)  #sub graph of 3R & 3C
ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x,y)
ax[1,1].pie(x)
ax[1,2].barh(x,y)
# ax[2,0].plot(x,y)
# ax[2,1].pie(x)
# ax[2,2].barh(x,y)
plt.savefig('Multiple Plots.png',format='png')
plt.show()
```

### Generating Plot within a Plot

```python
import numpy as np
x = np.arange(1,11)
y = x**2
print(x,y)
fig2 = plt.figure()

#Creating main axes - values b/w 0&1
axes_main = fig2.add_axes([0.1,0.1,0.9,0.9]) #[L,B,W,H]

#Creating insert axes - values b/w 0&1
axes_inset = fig2.add_axes([0.5,0.4,0.2,0.3])

#Change the numbers in axes_inset and see the difference
#Main figure
axes_main.plot(x, y, 'b')
axes_main.set_xlabel('X_label_main')
axes_main.set_ylabel('Y_label_main')
axes_main.set_title('Title_main')

#Inset figure
axes_inset.plot(y, x, 'r')
axes_inset.set_xlabel('X_label_inset')
axes_inset.set_ylabel('Y_label_inset')
axes_inset.set_title('Title_inset')

plt.show()
```

```python
# Setting Limits and ticks
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  # L,B,W,H
ax.set_xlim([2,6])
#ax.set_ylim([15,65])
ax.set_xticks([1,3,5])
#ax.set_yticks([9,25])
ax.plot(x,y)
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot([1,2,3,4,5],[3,5,7,1,9], color='red')
plt.show()
```

```python
x = np.arange(1,6)
y = x**2
print(x,y)
```

```python
plt.bar(x,y,width=[0.2,0.1,0.3,0.4,0.3],color=['cyan','red','green','yellow','violet'])
plt.show()
```

```python
#plt.labels = ['A','B','C','D','E']
plt.pie(x,[0,0.4,0,0,0],labels = ['A','B','C','D','E'])
plt.show()
```

```python
#Line Chart
plt.plot(x,y,lw=4,ls='-.',color='yellow',marker='p',mew=35,mfc='green',mec='cyan')
plt.show()
```
